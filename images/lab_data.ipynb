## Moving images

```python
# Function to move n random files from src_route to dst_route

def move_random_files(src_route, dst_route, n):
    images_list = os.listdir(src_route)

    if n > len(images_list):
        print(f"Error: ...")
        return

    os.makedirs(dst_route, exist_ok=True)
    selected = random.sample(images_list, n)

    for file in selected:
        shutil.move(os.path.join(src_route, file), os.path.join(dst_route, file))
        print(f"Movido: {file}")
```

## Deleting extra images

```python
def delete_random_files(dict_route, images_list, wanted_amount):

    if wanted_amount >= len(images_list):
        print(f"Error: Solicitaste eliminar {wanted_amount} archivos, pero solo hay {len(images_list)} disponibles!")
    else:
        deleting_amount = len(images_list) - wanted_amount
        random_files = random.sample(images_list, deleting_amount)

        for file in random_files:

            complete_route = os.path.join(dict_route, file)
            os.remove(complete_route)

            print(f"Eliminado: {file}")
            
        print(f"\nSe eliminaron {deleting_amount} archivos aleatorios.")
```
